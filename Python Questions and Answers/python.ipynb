Answer:

int: Represents whole numbers (positive or negative) without decimal points. Example: 5, -10.

float: Represents real numbers (positive or negative) with decimal points. Example: 3.14, -0.5.

* 14. What is the difference between / and //?
Answer:

/: Performs normal division and returns a float result.

//: Performs floor division and returns an integer result (rounded down to the nearest whole number).

> print(10 / 3)   # 3.333... (Float division)

> print(10 // 3)  # 3 (Floor division)

* 17. What are logical operators in Python?
Answer:
Logical operators are used to combine conditional statements. The logical operators in Python are:

and: Returns True if both conditions are true.

or: Returns True if at least one condition is true.

not: Reverses the result of a condition.

> x = 5

> y = 10

> print(x > 0 and y > 0)  # True (Both conditions are true)

> print(x > 0 or y < 0)   # True (At least one condition is true)

> print(not x > 0)        # False (Reverses the condition)

* 18. What is the purpose of the in operator?
Answer:
The in operator is used to check if a value exists in a sequence (e.g., list, tuple, string, etc.). It returns True if the value is found, otherwise False.

> fruits = ["apple", "banana", "cherry"]

> print("banana" in fruits)  # True

> print("mango" in fruits)   # False
